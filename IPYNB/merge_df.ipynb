```python
import matplotlib.pyplot as plt
import numpy as np
segments = 5

people = df['Exercise']
percentages = (np.random.randint(5,20, (len(people), segments)))
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
y_pos = np.arange(len(people))
colors ='rgwm'
left = np.zeros(len(people))
patch_handles = []
for i, d in enumerate(df):
    patch_handles.append(ax.barh(y_pos, d, color=colors[i%len(colors)], align='center', left=left))
    left += d

# search all of the bar segments and annotate
for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5*patch.get_width() + bl[0]
        y = 0.5*patch.get_height() + bl[1]
        ax.text(x,y, "%d%%" % (percentages[i,j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('BMI bin')
plt.show()
```

```python
y1 = np.array(list(df[df['Exercise'] == 'None']['BMIbin']))
y2 = list(df[df['Exercise'] == 'Moderate']['BMIbin'])
y3 = list(df[df['Exercise'] == 'Heavy']['BMIbin'])

# memo of sample number
snum = len(y1)+len(y2)+len(y3)
x = ["Normal","Obesity","Overweight","Underweight","Very Severely Underweight"]
# normalization
y1 = y1/snum*100
y2 = y2/snum*100
y3 = y3/snum*100
plt.figure(figsize=(4,3))

# stack bars
plt.bar(x, y1, label='y1')
plt.bar(x, y2 ,bottom=y1,label='y2')
plt.bar(x, y3 ,bottom=y1+y2,label='y3')

# add text annotation corresponding to the percentage of each data.
for xpos, ypos, yval in zip(x, y1/2, y1):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")
for xpos, ypos, yval in zip(x, y1+y2/2, y2):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")
for xpos, ypos, yval in zip(x, y1+y2+y3/2, y3):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")
# add text annotation corresponding to the "total" value of each bar
for xpos, ypos, yval in zip(x, y1+y2+y3, snum):
    plt.text(xpos, ypos, "N=%d"%yval, ha="center", va="bottom")

plt.ylim(0,110)

plt.legend(bbox_to_anchor=(1.01,0.5), loc='center left')
plt.savefig('normalized_stacked_barplot_with_number.png', bbox_inches='tight', pad_inches=0.02)
```
