```python
j = []
for i in df.itertuples(): 
    if(i[2]<70):
        j.append('Definite feeble mindedness')
    elif(i[2]<80):
        j.append('Borderline deficiency')
    elif(i[2]<90):
        j.append('Dullness')
    elif(i[2]<110):
        j.append('Normal or average')
    elif(i[2]<120):
        j.append('Superior intelligence')
    elif(i[2]<141):
        j.append('Very superior intelligence')
    else:
        j.append('Genius or near genius')
```

```python
df['IQScore Category'] = j
df.groupby('IQScore Category')['Suicide Possibility'].value_counts()
```

```python
df1 = df[df['IQScore Category']!='Genius or near genius'].groupby('IQScore Category')['Suicide Possibility'].value_counts()
import matplotlib 
%matplotlib inline
df1.unstack().plot(kind = 'bar',stacked=True,figsize=(8,5))
```
